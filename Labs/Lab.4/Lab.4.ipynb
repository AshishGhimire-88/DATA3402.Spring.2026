1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value, current_value=0):
        self.max_value= max_value
        self.current_value= current_value
        if self.current_value>self.max_value:
            print("Initial value can't be greater than maximum value.")

    def increment(self, increment_value=1):
        if self.current_value >= self.max_value:
            print(f"Counter reached its maximum value of {self.max_value}")
        else:
            if self.current_value+increment_value<=self.max_value:
                self.current_value+=increment_value
            
            else: 
                print(f"If incremented by {increment_value}, counter Will exceed its maximum value of {self.max_value}")
    
    def reset(self, reset_to=0):
        if reset_to>self.max_value:
            print(f"Reset value can't be larger than {self.max_value}")
        else:
            self.current_value= reset_to
    
    def __str__(self):
        return f"Current Counter Value: {self.current_value}"
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members. 

```python
import math

class Circle0():
    def __init__(self, radius=None, x=0, y=0):
        if radius is None:
            raise ValueError("Please enter a radius for the circle.")
        
        self.__radius= radius
        self.__x= x
        self.__y= y
    
    @property       #This turns the function into a managable attribute.
    def area(self):
        return math.pi * (self.__radius**2)
    
    @property
    def perimeter(self):
        return 2* math.pi * self.__radius
    
    @property
    def x(self):
        return self.__x
    
    @property
    def y(self):
        return self.__y
    
    def __str__(self):
        return f"Circle with radius {self.__radius} and centre at {(self.__x, self.__y)}"
```

```python
#Implementing Triangle Class
class Triangle3(Shape3):
    def __init__(self, a, b, c, x=0, y=0):
        #validate the triangle:
        if a+b<=c or b+c<=a or c+a<=b:
            raise ValueError("Enter a valid triangle lengths!")
        super().__init__(x=x, y=y)
        self._a= a
        self._b=b
        self._c=c
    
    @property
    def area(self):
        s = self.perimeter / 2
        return math.sqrt(s*(s-self._a)*(s-self._b)*(s-self._c))
    
    @property
    def perimeter(self):
        return self._a + self._b + self._c
    
    def _get_vertices(self):
        p1= (self._x, self._y)
        p2= (self._x+self._c, self._y)
        cosangle=(self._b**2 + self._c**2 - self._a**2)/(2*self._b*self._c)
        sinangle = math.sqrt(1 - cosangle**2)
        p3= (self._x+self._b*cosangle, self._y+self._b*sinangle)
        return p1,p2,p3

    def points(self, max_points=16):
        if max_points>16 or max_points<=3:            #Added max_point<=4 for later comparision easiness
            raise ValueError("Max point has to be less than 16 and more than 4")
        pts=list()
        p1,p2,p3= self._get_vertices()

        pts.extend([p1, p2, p3])
    
        while len(pts) < max_points:
            new_pts = []
            for i in range(len(pts)):
                p1 = pts[i]
                p2 = pts[(i+1) % len(pts)]
                midpoint = (((p1[0]+p2[0])/2), (p1[1]+p2[1])/2)
                new_pts.append(p1)
                new_pts.append(midpoint)
                if len(new_pts) >= max_points:
                    break
            pts = new_pts
        return pts[:max_points]
    

    def point_test(self, x_test, y_test):       #Follows the 3 Area calculation formula: A1+A2+A3<= Total Area
        p1,p2,p3= self._get_vertices()

        def calculate_area(p1,p2,p3):
            return abs(1/2*(p1[0]*(p2[1]-p3[1])+ p2[0]*(p3[1]-p1[1])+ p3[0]*(p1[1]-p2[1])))
        
        total_area= calculate_area(p1, p2, p3)
        area1= calculate_area(p1,p2,(x_test, y_test))
        area2= calculate_area(p2,p3,(x_test, y_test))
        area3= calculate_area(p3,p1,(x_test, y_test))
        return abs((area1 + area2 + area3)- total_area) <1e-9

    def __str__(self):
        return f"Triangle with sides: {self._a}, {self._b} & {self._c} and one vertex at{(self.x, self.y)}"
```

```python
# Implementing Parent Class: Shape
class Shape():
    def __init__(self, x=0, y=0):
        self._x= x
        self._y=y
    
    @property
    def area(self):
        raise NotImplementedError("Subclass Should Implement Area.")        #Virtul not implemented method
    
    @property
    def perimeter(self):
        raise NotImplementedError("Subclass Should Implement Perimeter.")       #Virtual not implemented method
    
    @property
    def x(self):
        return self._x
    
    @property
    def y(self):
        return self._y
    
    def points(self):
        raise NotImplementedError("Subclass Should Implement Points.")
    
    def point_test(self):
        raise NotImplementedError("Subclass Should Implement the Test.")
    
    def check_overlapping(self, other, max_points=16):
        self_points= self.points(max_points)
        other_points= other.points(max_points)
        for pts in self_points:
            if other.point_test(pts[0], pts[1]):
                return True
        for pts in other_points:
            if self.point_test(pts[0], pts[1]):
                return True
        return False
```

```python
class Circle(Shape):
    def __init__(self, radius=None, x=0, y=0):
        super().__init__(x=x, y=y)
        if radius is None:
            raise ValueError("Please enter a radius for the circle.")
        
        self._radius= radius
    
    @property       #This turns the function into a managable attribute.
    def area(self):
        return math.pi * (self._radius**2)
    
    @property
    def perimeter(self):
        return 2* math.pi * self._radius
    
    def points(self, max_point=16):
        if max_point > 16 or max_point <= 4:
            raise ValueError("Max point has to be less than 16 and more than 4")
        pts = []
        for i in range(max_point):
            angle = 2 * math.pi * i / max_point
            x = self._x + self._radius * math.cos(angle)
            y = self._y + self._radius * math.sin(angle)
            pts.append((x, y))
        return pts
    
    def point_test(self, x1, y1):
        return math.sqrt((x1-self._x)**2 + (y1-self._y)**2)<= self._radius
    
    def __str__(self):
        return f"Circle with radius {self._radius} and centre at {(self.x, self.y)}"
```

```python
#Implementing Triangle Class
class Triangle(Shape):
    def __init__(self, a, b, c, x=0, y=0):
        #validate the triangle:
        if a+b<=c or b+c<=a or c+a<=b:
            raise ValueError("Enter a valid triangle lengths!")
        super().__init__(x=x, y=y)
        self._a= a
        self._b=b
        self._c=c
    
    @property
    def area(self):
        s = self.perimeter / 2
        return math.sqrt(s*(s-self._a)*(s-self._b)*(s-self._c))
    
    @property
    def perimeter(self):
        return self._a + self._b + self._c
    
    def _get_vertices(self):
        p1= (self._x, self._y)
        p2= (self._x+self._c, self._y)
        cosangle=(self._b**2 + self._c**2 - self._a**2)/(2*self._b*self._c)
        sinangle = math.sqrt(1 - cosangle**2)
        p3= (self._x+self._b*cosangle, self._y+self._b*sinangle)
        return p1,p2,p3

    def points(self, max_points=16):
        if max_points>16 or max_points<=3:            #Added max_point<=4 for later comparision easiness
            raise ValueError("Max point has to be less than 16 and more than 4")
        pts=list()
        p1,p2,p3= self._get_vertices()

        pts.extend([p1, p2, p3])
    
        while len(pts) < max_points:
            new_pts = []
            for i in range(len(pts)):
                p1 = pts[i]
                p2 = pts[(i+1) % len(pts)]
                midpoint = (((p1[0]+p2[0])/2), (p1[1]+p2[1])/2)
                new_pts.append(p1)
                new_pts.append(midpoint)
                if len(new_pts) >= max_points:
                    break
            pts = new_pts
        return pts[:max_points]
    

    def point_test(self, x_test, y_test):       #Follows the 3 Area calculation formula: A1+A2+A3<= Total Area
        p1,p2,p3= self._get_vertices()

        def calculate_area(p1,p2,p3):
            return abs(1/2*(p1[0]*(p2[1]-p3[1])+ p2[0]*(p3[1]-p1[1])+ p3[0]*(p1[1]-p2[1])))
        
        total_area= calculate_area(p1, p2, p3)
        area1= calculate_area(p1,p2,(x_test, y_test))
        area2= calculate_area(p2,p3,(x_test, y_test))
        area3= calculate_area(p3,p1,(x_test, y_test))
        return abs((area1 + area2 + area3)- total_area) <1e-9

    def __str__(self):
        return f"Triangle with sides: {self._a}, {self._b} & {self._c} and one vertex at{(self.x, self.y)}"
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, row, col, height, char='*'):
        for r in range(row, row + height):
            self.set_pixel(r, col, char)

    def h_line(self, row, col, length, char='*'):
        for c in range(col, col + length):
            self.set_pixel(row, c, char)
            
    
    #NOTE: Fixed the line method using LLM to meet the requirements efficiently.
    def line(self, r1, c1, r2, c2, char='*'):
        dr = r2 - r1
        dc = c2 - c1
        steps = max(abs(dr), abs(dc))

        if steps == 0:
            self.set_pixel(r1, c1, char)
            return

        r_inc = dr / steps
        c_inc = dc / steps

        r = r1
        c = c1

        for _ in range(steps + 1):
            self.set_pixel(int(round(r)), int(round(c)), char)
            r += r_inc
            c += c_inc
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
class CompoundShape(Shape):
    def __init__(self):
        super().__init__()
        self.shapes= list()
    
    def add_shape(self, shape):
        self.shapes.append(shape)
        
    def remove_shape(self, shape):
        self.shape.remove(shape)
    
    @property
    def perimeter(self):
        return sum(shape.perimeter for shape in self.shapes)
    
    @property
    def area(self):
        return sum(shape.area for shape in self.shapes)
    
    def points(self, max_points=16):
        pts=[]
        for shape in self.shapes:
            pts.extend(shape.points(max_points))
        return pts[:max_points]
    
    def point_test(self, x1, y1):
        return any(shape.point_test(x1, y1) for shape in self.shapes)
    
    def paint(self, canvas, char="*"):
        for shape in self.shapes:
            shape.paint(canvas, char)
    
    def __str__(self):
        return f"CompoundShape with {len(self.shapes)} shpaes in it"
```

```python
#Implementing Triangle Class
class Triangle(Shape):
    def __init__(self, a, b, c, x=0, y=0):
        #validate the triangle:
        if a+b<=c or b+c<=a or c+a<=b:
            raise ValueError("Enter a valid triangle lengths!")
        super().__init__(x=x, y=y)
        self._a= a
        self._b=b
        self._c=c
    
    @property
    def area(self):
        s = self.perimeter / 2
        return math.sqrt(s*(s-self._a)*(s-self._b)*(s-self._c))
    
    @property
    def perimeter(self):
        return self._a + self._b + self._c
    
    def _get_vertices(self):
        p1= (self._x, self._y)
        p2= (self._x+self._c, self._y)
        cosangle=(self._b**2 + self._c**2 - self._a**2)/(2*self._b*self._c)
        sinangle = math.sqrt(1 - cosangle**2)
        p3= (self._x+self._b*cosangle, self._y+self._b*sinangle)
        return p1,p2,p3

    def points(self, max_points=16):
        if max_points<=3:            #Added max_point<=4 for later comparision easiness
            raise ValueError("Max point has to be more than 4")
        pts=list()
        p1,p2,p3= self._get_vertices()

        pts.extend([p1, p2, p3])
    
        while len(pts) < max_points:
            new_pts = []
            for i in range(len(pts)):
                p1 = pts[i]
                p2 = pts[(i+1) % len(pts)]
                midpoint = (((p1[0]+p2[0])/2), (p1[1]+p2[1])/2)
                new_pts.append(p1)
                new_pts.append(midpoint)
                if len(new_pts) >= max_points:
                    break
            pts = new_pts
        return pts[:max_points]
    

    def point_test(self, x_test, y_test):       #Follows the Area method: A1+A2+A3<= Total Area
        p1,p2,p3= self._get_vertices()

        def calculate_area(p1,p2,p3):
            return abs(1/2*(p1[0]*(p2[1]-p3[1])+ p2[0]*(p3[1]-p1[1])+ p3[0]*(p1[1]-p2[1])))
        
        total_area= calculate_area(p1, p2, p3)
        area1= calculate_area(p1,p2,(x_test, y_test))
        area2= calculate_area(p2,p3,(x_test, y_test))
        area3= calculate_area(p3,p1,(x_test, y_test))
        return abs((area1 + area2 + area3)- total_area) <1e-9
    
    def paint(self, canvas, char='o'):
        p1, p2, p3 = self._get_vertices()
        r1, c1 = int(p1[1]), int(p1[0])
        r2, c2 = int(p2[1]), int(p2[0])
        r3, c3 = int(p3[1]), int(p3[0])

        canvas.h_line(r1, c1, c2 - c1, char=char)  # base - horizontal so use h_line
        canvas.line(r1, c1, r3, c3, char=char)     # left side
        canvas.line(r2, c2, r3, c3, char=char)     # right side

    def __repr__(self):
        return f"Triangle(a={self._a}, b={self._b}, c={self._c}, x={self._x}, y={self._y})"

    def __str__(self):
        return f"Triangle with sides: {self._a}, {self._b} & {self._c} and one vertex at{(self.x, self.y)}"
```

```python
class RasterDrawing:
    def __init__(self, canvas):
        self.canvas = canvas
        self.shapes = []
    
    def add_shape(self, shape):
        self.shapes.append(shape)
    
    def remove_shape(self, shape):
        self.shapes.remove(shape)
    
    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(self.canvas)
    
    def __str__(self):
        return f"RasterDrawing with {len(self.shapes)} shapes"
    
    def save(self, filename):
        f = open(filename, "w")
        for shape in self.shapes:
            f.write(repr(shape) + "\n")
        f.close()

def raster_loader(filename):
    f = open(filename, "r")
    drawing = RasterDrawing(Canvas(40, 40))
    for line in f.readlines():
        drawing.add_shape(eval(line))
    f.close()
    return drawing
```

```python
canvas = Canvas(40, 40)
drawing = RasterDrawing(canvas)
drawing.add_shape(Rectangle(length=5, width=3, x=2, y=2))
drawing.add_shape(Circle(radius=3, x=15, y=15))
drawing.add_shape(Triangle(3, 4, 5, x=5, y=5))

print("Original Drawing:")
drawing.paint()
canvas.display()

# Save The DRawing
drawing.save("my_drawing.txt")
print("\nFile Contents:")
with open("my_drawing.txt", "r") as f:
    print(f.read())

# Load the Save file
drawing2 = raster_loader("my_drawing.txt")
print("Loaded Drawing: ")
drawing2.paint()
drawing2.canvas.display()
```
